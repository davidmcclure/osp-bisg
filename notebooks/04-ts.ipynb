```python
import warnings

import pandas as pd
import numpy as np
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
warnings.simplefilter('ignore')
```

```python
pd.options.display.max_rows = 200
```

```python
ng1_counts = pd.read_json('../data/ng1-year-counts.json/part-00000-b7fd6866-48db-4aac-919e-02782ba0092f-c000.json', lines=True)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='year', y='count', data=ng1_counts, palette='viridis')

plt.rc('axes', labelsize=16)
plt.rc('axes', titlesize=22)
plt.title('Total number of words')
plt.xlabel('Year')
plt.ylabel('Words')
```

```python
def plot_kw(path, title):
    
    kw_counts = pd.read_json(path, lines=True)
    kw_counts['rel_count'] = kw_counts['count'] / ng1_counts['count']

    plt.figure(figsize=(10,5))
    sns.barplot(x='year', y='rel_count', data=kw_counts, palette='viridis')
    
    plt.rc('axes', labelsize=16)
    plt.rc('axes', titlesize=22)
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('Relative count')
```

```python
he_counts = pd.read_json('../data/he.json/part-00000-11ab06fa-446c-4a98-9d54-7acd7c334b4c-c000.json', lines=True)
she_counts = pd.read_json('../data/she.json/part-00000-4ea74d9a-2940-4889-b2af-9f8172a7fb63-c000.json', lines=True)
he_counts['ratio'] = he_counts['count'] / she_counts['count']
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='year', y='ratio', data=he_counts, palette='viridis')

plt.rc('axes', labelsize=16)
plt.rc('axes', titlesize=22)
plt.title('Ratio of "he" / "she"')
plt.xlabel('Year')
plt.ylabel('Ratio')
```
